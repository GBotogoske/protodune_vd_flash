```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal


list_ch =[1050,1051,1060,1061,1070,1071,1080,1081,2030,2031,2040,2041,2070,2071,2080,2081]
list_spe=[39357,39357,39358,39359,39360,39360,39361,39361,39360,39360,39360,39360,39365,39365,39366,39366]

n_ch = len(list_ch)

charge=[]

for ch in range(n_ch):
    file="/home/gabriel/Documents/protodune/protodune_vd/light_analysis/waveforms/"+f"{list_spe[ch]}/template_{list_ch[ch]}.txt"
    template = np.loadtxt(file)
    dtemplate = np.diff(template)
    index_max=np.argmax(dtemplate)

    print(index_max)
    plt.plot(template)
    plt.axvline(index_max,linestyle="--",color="red")
    plt.axvline(index_max+200,linestyle="--",color="red")
    charge.append(np.sum(template[index_max:index_max+100]))
    
    
```

```python
np.savetxt("/home/gabriel/Documents/protodune/protodune_vd/light_analysis/charge_spe.txt", charge, fmt="%.6f")
```
