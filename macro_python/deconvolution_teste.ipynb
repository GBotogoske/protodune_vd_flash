```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

from scipy.fft import fft, fftshift

def gaus(x, sigma=60): #sigma = sqrt(sqrt(2)) * cutoff
    return np.exp(-(x)**2/2/sigma**2)

N=2048
_x =  np.fft.fftfreq(N) * N #np.linspace(0, 1024, 1024, endpoint=False)
filter_gaus = [ gaus(x,60) for x in _x]

list_run = [39510,39511,39512,39514,39515,39516,39517,39518,39519,39521,39522,39523,39525,39526,39528,39529]
list_ch =[1050,1051,1060,1061,1070,1071,1080,1081,2030,2031,2040,2041,2070,2071,2080,2081]
list_spe=[39357,39357,39358,39359,39360,39360,39361,39361,39360,39360,39360,39360,39365,39365,39366,39366]

for i in range(len(list_run)):
    for j in range(len(list_ch)):

        template_file="/home/gabriel/Documents/protodune/protodune_vd/light_analysis/waveforms/"+f"{list_spe[j]}"+"/waveform_"+f"{list_ch[j]}"+"_filtered.txt"
        template_file_nofiltered= "/home/gabriel/Documents/protodune/protodune_vd/light_analysis/waveforms/"+f"{list_spe[j]}"+"/waveform_"+f"{list_ch[j]}"+".txt"

        data_template = np.loadtxt(template_file)
        t_template = data_template[:, 0]
        s_template = data_template[:, 1]

        data_template_nofiltered = np.loadtxt(template_file_nofiltered)
        t_template_nofiltered = data_template_nofiltered[:, 0]
        s_template_nofiltered = data_template_nofiltered[:, 1]

        signal_file="/home/gabriel/Documents/protodune/protodune_vd/light_analysis/waveforms/"+f"{list_run[i]}"+"/waveform_modamedia"+f"{list_ch[j]}"+".txt"


        data = np.loadtxt(signal_file)
        t = data[:, 0]
        s = data[:, 1]

        template_fft=np.fft.fft(s_template,2048)  
        
        s_fft=np.fft.fft(s,2048)
        x_fft=s_fft/template_fft
        for k, _ in enumerate(x_fft):
            x_fft[k] *= filter_gaus[k]

        x=np.real(np.fft.ifft(x_fft))
        x=np.roll(x,253)[0:1024]


        plt.plot(x)
        plt.grid()
        plt.savefig("/home/gabriel/Documents/protodune/protodune_vd/light_analysis/waveforms/"+f"{list_run[i]}"+"/waveform_"+f"{list_ch[j]}_filt_dec"+".png")
        np.savetxt("/home/gabriel/Documents/protodune/protodune_vd/light_analysis/waveforms/"+f"{list_run[i]}"+"/waveform_"+f"{list_ch[j]}_filt_dec"+".txt", np.column_stack((t, x)), fmt="%.6e", delimiter="\t")
        plt.close()

    
```
